"""У тебя есть группа студентов. Нужно смоделировать данные и провести аналитику.
 1. Создай DataFrame со столбцами:
 • Имя (10–15 случайных имён),
 • Возраст (от 18 до 25),
 • Группа (например: “Физика”, “Математика”, “Информатика”),
 • Оценка за экзамен (от 2 до 5),
 • Посещаемость (%) (от 50 до 100).
 2. Добавь колонку «Статус»:
 • Если Оценка ≥ 4 и Посещаемость ≥ 75 → «Успешный»,
 • Если Оценка ≥ 3 и Посещаемость ≥ 60 → «Средний»,
 • Иначе → «Неуспевающий».
 3. Сделай группировку по «Группа» и найди:
 • среднюю оценку,
 • среднюю посещаемость,
 • количество «успешных» студентов.
 4. Красиво оформи вывод:
 • подсвети «успешных» студентов зелёным,
 • «неуспевающих» — красным,
 • добавь градиент по посещаемости.
 5. Найди:
 • студента с лучшей посещаемостью,
 • группу с самым высоким средним баллом. """

```python
import random

import pandas as pd
k = int(input("Введите количество учеников в группе"))
names = ["Мария","Никита","Саша","Петя","Ваня","Катя","Алиса","Арина","Алина","Анжела","Георгий","Савелий","Настя","Кира"]
groups = ["Математика","Физика","Информатика","Английский","Физ-ра","Немецкий","Философия","Психология","Русский язык"]
name = [random.choice(names) for i in range(k)]
statuses = []
print(name)
vozr = [random.randint(18,25) for i in range(k)]
group = [random.choice(groups) for i in range(k)]
mark = [random.randint(2,5) for i in range(k)]
attendance = [random.randint(50,100) for i in range(k)]
df1 = pd.DataFrame({"Имя":name,"Возраст":vozr,"Группа":group,"Оценка":mark,"Посещаемость":attendance})

for i in range(len(df1)):

    if df1.loc[i,"Посещаемость"] >= 75 and df1.loc[i,"Оценка"] >= 4:
        statuses.append("Успешный")
    elif df1.loc[i,"Посещаемость"] >= 60 and df1.loc[i,"Оценка"] >= 3:
        statuses.append("Средний")
    else:
        statuses.append("Неуспевающий")
df1["Статус"] = statuses
print(df1)

x = df1.groupby("Группа")["Возраст"].mean()
print(x," - Средний возраст по группам")
m = df1.groupby("Группа")["Посещаемость"].mean()
print(m," - Средняя посещаемость по группам")
c = (df1["Статус"] == "Успешный").sum()
print(c," - Количество успешных учеников")

maxmark=df1["Оценка"].max()
student=df1[df1["Оценка"]==maxmark]
print("Студент с наивысшей оценкой:")
print(student[["Имя","Группа","Оценка"]])

sr=df1.groupby("Группа")["Оценка"].mean()
luchsh=sr[sr==sr.max()]
print("Группа с наивысшей оценкой:")
print(luchsh)


styleddf2= df1.style.map(
        lambda v: 'background-color: lightgreen; text-decoration: underline' if v == 'Успешный' else '', subset=['Статус']
    )
styleddf2
```
